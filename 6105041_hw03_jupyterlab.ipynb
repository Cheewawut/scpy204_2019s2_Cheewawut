```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('COVID-19-geographic-disbtribution-worldwide-2020-04-12.csv')
#change data type of dateRep to datetime
data['dateRep'] = pd.to_datetime(data['dateRep'],format = "%d/%m/%Y")
#sorting data by dateRep
data.sort_values(by='dateRep',inplace = True) 
```

```python
############ 3
```

```python
TH = data[data.countriesAndTerritories == "Thailand"] 
US = data[data.countriesAndTerritories == "United_States_of_America"] 
UK = data[data.countriesAndTerritories == "United_Kingdom"] 
JP = data[data.countriesAndTerritories == "Japan"]
IT = data[data.countriesAndTerritories == "Italy"]
CN = data[data.countriesAndTerritories == "China"]
KR = data[data.countriesAndTerritories == "South_Korea"]
ES = data[data.countriesAndTerritories == "Spain"]
FR = data[data.countriesAndTerritories == "France"]
DE = data[data.countriesAndTerritories == "Germany"]
```

```python
popTH = TH.popData2018.max()
popUS = US.popData2018.max()
popUK = UK.popData2018.max()
popJP = JP.popData2018.max()
popIT = IT.popData2018.max()
popCN = CN.popData2018.max()
popKR = KR.popData2018.max()
popES = ES.popData2018.max()
popFR = FR.popData2018.max()
popDE = DE.popData2018.max()

```

```python
#collecting cases and deaths of TH
scaseTH = TH.cases
sdeathTH = TH.deaths
```

```python
#sum cases and deaths of TH
caseTH = scaseTH.sum()
deathTH = sdeathTH.sum()
```

```python
pcaseTH = caseTH*100/popTH
pdeathTH = deathTH*100/caseTH
```

```python
print( "The percentage of infected people is ", pcaseTH )
print( "The percentage of dead people is ", pdeathTH )
```

```python
#detemine the cummulative patients from each country and date.
dateTH = TH.dateRep
ccaseTH = TH.cases.cumsum()

dateUS = US.dateRep
ccaseUS = US.cases.cumsum()

dateUK = UK.dateRep
ccaseUK = UK.cases.cumsum()

dateJP = JP.dateRep
ccaseJP = JP.cases.cumsum()

dateIT = IT.dateRep
ccaseIT = IT.cases.cumsum()

dateCN = CN.dateRep
ccaseCN = CN.cases.cumsum()

dateKR = KR.dateRep
ccaseKR = KR.cases.cumsum()

dateES = ES.dateRep
ccaseES = ES.cases.cumsum()

dateFR = FR.dateRep
ccaseFR = FR.cases.cumsum()

dateDE = DE.dateRep
ccaseDE = DE.cases.cumsum()
```

```python
plt.plot_date(dateTH,ccaseTH,'-',c='crimson',label='Thailand')
plt.plot_date(dateUS,ccaseUS,'-',c='green',label='United State of America')
plt.plot_date(dateUK,ccaseUK,'-',c='blue',label='United Kingdom')
plt.plot_date(dateJP,ccaseJP,'-',c='orange',label='Japan')
plt.plot_date(dateIT,ccaseIT,'-',c='darkolivegreen',label='Italy')
plt.plot_date(dateCN,ccaseCN,'-',c='black',label='China')
plt.plot_date(dateKR,ccaseKR,'-',c='yellow',label='South Korean')
plt.plot_date(dateFR,ccaseFR,'-',c='lightskyblue',label='France')
plt.plot_date(dateES,ccaseES,'-',c='orchid',label='Spain')
plt.plot_date(dateDE,ccaseDE,'-',c='c',label='Germany')

plt.style.use('seaborn')
plt.title("The Commulative Patients of Covid-19")
plt.xlabel("Year-Month-Date")
plt.ylabel("A Number of Patients")
plt.gcf().autofmt_xdate() #rotating date

plt.yscale('log')
plt.legend()
plt.show()
```

```python

pcaseUS = US.cases.sum()*100/popUS
pcaseUK = UK.cases.sum()*100/popUK
pcaseJP = JP.cases.sum()*100/popJP
pcaseIT = IT.cases.sum()*100/popIT
pcaseCN = CN.cases.sum()*100/popCN
pcaseKR = KR.cases.sum()*100/popKR
pcaseES = ES.cases.sum()*100/popES 
pcaseFR = FR.cases.sum()*100/popFR
pcaseDE = DE.cases.sum()*100/popDE

pdeathUS = US.deaths.sum()*100/US.cases.sum()
pdeathUK = UK.deaths.sum()*100/UK.cases.sum()
pdeathCN = CN.deaths.sum()*100/CN.cases.sum()
pdeathJP = JP.deaths.sum()*100/JP.cases.sum()
pdeathIT = IT.deaths.sum()*100/IT.cases.sum()
pdeathKR = KR.deaths.sum()*100/KR.cases.sum()
pdeathES = ES.deaths.sum()*100/ES.cases.sum()
pdeathFR = FR.deaths.sum()*100/FR.cases.sum()
pdeathDE = DE.deaths.sum()*100/DE.cases.sum()

```

```python
plt.scatter(pcaseTH,pdeathTH,c='crimson',s=popTH/10**6,alpha=0.3,linewidth=2,label='Thailand')
plt.scatter(pcaseCN,pdeathCN,c='green',s=popCN/10**6,alpha=0.3,linewidth=2,label='China')
plt.scatter(pcaseUS,pdeathUS,c='blue',s=popUS/10**6,alpha=0.3,linewidth=2,label='United State')
plt.scatter(pcaseUK,pdeathUK,c='orange',s=popUK/10**6,alpha=0.3,linewidth=2,label='United Kingdom')
plt.scatter(pcaseJP,pdeathJP,c='darkolivegreen',s=popJP/10**6,alpha=0.3,linewidth=2,label='Japan')
plt.scatter(pcaseIT,pdeathIT,c='black',s=popIT/10**6,alpha=0.3,linewidth=2,label='Italy')
plt.scatter(pcaseKR,pdeathKR,c='yellow',s=popKR/10**6,alpha=0.3,linewidth=2,label='South Korea')
plt.scatter(pcaseES,pdeathES,c='lightskyblue',s=popES/10**6,alpha=0.3,linewidth=2,label='Spain')
plt.scatter(pcaseFR,pdeathFR,c='orchid',s=popFR/10**6,alpha=0.3,linewidth=2,label='France')
plt.scatter(pcaseDE,pdeathDE,c='c',s=popDE/10**6,alpha=0.3,linewidth=2,label='Germany')

plt.legend(loc = (1.05,0.1),title='Countries',labelspacing=1,fontsize=12)
plt.title("The Relationship of infected and dead people from Covid-19",fontsize=15)
plt.xlabel("The percentage of Infected People per Poppulation",fontsize=15)
plt.ylabel("The percentage of dead people per infected people",fontsize=14)

plt.show()
```
